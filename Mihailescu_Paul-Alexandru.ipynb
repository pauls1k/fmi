```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os.path

from sklearn import metrics
from sklearn.svm import LinearSVC
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.metrics import f1_score, confusion_matrix, classification_report, accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.pipeline import Pipeline
```

```
# Citesc fisierele

train_samples_df = pd.read_csv(os.path.join(DATA_PATH, 'train_samples.txt'), sep = '\t', names=['id', 'document'])
train_labels_df = pd.read_csv(os.path.join(DATA_PATH, 'train_labels.txt'), sep = '\t', names=['id', 'label'])

validation_samples_df = pd.read_csv(os.path.join(DATA_PATH, 'validation_samples.txt'), sep = '\t', names=['id', 'document'])
validation_labels_df = pd.read_csv(os.path.join(DATA_PATH, 'validation_labels.txt'), sep = '\t', names=['id', 'label'])

test_samples_df = pd.read_csv(os.path.join(DATA_PATH, 'test_samples.txt'), sep = '\t', names=['id', 'document'])
```

```
# Pun datele citite anterior intr-o lista

train_data = np.array(train_samples_df['document'])
train_labels = np.array(train_labels_df['label'])

validation_data = np.array(validation_samples_df['document'])
validation_labels = np.array(validation_labels_df['label'])

test_data = np.array(test_samples_df['document'])

print(train_data)
```

```
# Equivalent to CountVectorizer followed by TfidfTransformer.
# Cu TfidfVectorizer convertesc lista cu text intr-o matrice normalizata

v = TfidfVectorizer(lowercase = False, token_pattern = ('\S+'),ngram_range=(1,1)) # LOWERCASE ?????? NGRAM????????????


train_data_tfidf = v.fit_transform(train_data)

validation_data_tfidf = v.transform(validation_data)

test_data_tfidf = v.transform(test_data)


#print(train_data_tfid.shape)
#print(v.vocabulary_)
```

```
#Naive Bayes
mnb = MultinomialNB()
mnb.fit(train_data_tfidf, train_labels)

#LinearSVC
svc = LinearSVC()
svc.fit(train_data_tfidf, train_labels)

```

```
# EVALUAREA MODELELOR

#Naive Bayes
print('\nNB\n')
vali_pred_nb = mnb.predict(validation_data_tfidf)

print(classification_report(validation_labels, vali_pred_nb))
print(confusion_matrix(validation_labels, vali_pred_nb))

f1_macro = f1_score(validation_labels, vali_pred_nb, average='macro')
print(f1_macro)

#LinearSVC
print('\nSVC\n')
vali_pred_svc = svc.predict(validation_data_tfidf)

print(classification_report(validation_labels, vali_pred_svc))
print(confusion_matrix(validation_labels, vali_pred_svc))

f1_macro = f1_score(validation_labels, vali_pred_svc, average='macro')
print(f1_macro)
```

```
# GRID SEARCH

parametri_nb = {'alpha': [1e-1,1e-2,1e-3]}
parametri_svc = {'C': [100,10,5,4,3,2,1,0.8,0.5,0.2]}

grid_nb = GridSearchCV(MultinomialNB(),parametri_nb,verbose=1,cv=5)
grid_svc = GridSearchCV(LinearSVC(),parametri_svc,verbose=1,cv=5)

grid_nb.fit(train_data_tfidf, train_labels)
grid_svc.fit(train_data_tfidf, train_labels)

print(grid_nb.best_params_, grid_svc.best_params_)
print(grid_nb.best_score_,grid_svc.best_score_)

```

```
# Grid NB
print("\ngrid nb\n")
grid_vali_pred_nb = grid_nb.predict(validation_data_tfidf)

print(classification_report(validation_labels, grid_vali_pred_nb))
print(confusion_matrix(validation_labels, grid_vali_pred_nb))

f1_macro = f1_score(validation_labels, grid_vali_pred_nb, average='macro')
print(f1_macro)

# Grid SVC
print("\ngrid svc\n")
grid_vali_pred_svc = grid_svc.predict(validation_data_tfidf)

print(classification_report(validation_labels, grid_vali_pred_svc))
print(confusion_matrix(validation_labels, grid_vali_pred_svc))

f1_macro = f1_score(validation_labels, grid_vali_pred_svc, average='macro')
print(f1_macro)

```

```
# Rularea modelului pe datele de test

y_pred_nb = grid_nb.predict(test_data_tfidf)
print(y_pred_nb.shape)
print(test_data.shape)

y_pred_svc = grid_svc.predict(test_data_tfidf)
print(y_pred_svc.shape)
print(test_data.shape)

```

```
sol_data_nb = {
    'id': test_samples_df['id'],
    'label': y_pred_nb
}
solution_df = pd.DataFrame(data=sol_data_nb, dtype=np.int32, columns=['id', 'label'], copy=True)
sol_fpath = os.path.join(DATA_PATH.replace('date', 'solutii'), 'solutie_test_nb.txt')
solution_df.to_csv(sol_fpath, index=False, header=True)

sol_data_svc = {
    'id': test_samples_df['id'],
    'label': y_pred_svc
}
solution_df = pd.DataFrame(data=sol_data_svc, dtype=np.int32, columns=['id', 'label'], copy=True)
sol_fpath = os.path.join(DATA_PATH.replace('date', 'solutii'), 'solutie_test_svc.txt')
solution_df.to_csv(sol_fpath, index=False, header=True)
```
